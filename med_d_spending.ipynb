```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Load your actual medicare_d_spending DataFrame ---
try:
    medicare_d_spending = pd.read_csv(r'C:\Users\Faisal Askari\Downloads\DSD_PTD_RY24_P04_V10_DY22_BGM.csv')
    print("CSV loaded successfully!")
except FileNotFoundError:
    print("Error: The CSV file was not found. Please check the path and try again.")
    exit()

# --- 2. Define parameters ---
years = range(2018, 2023)
MANUFACTURER_NAME_COL = 'Mftr_Name' # Column for Manufacturer Name
BRAND_NAME_COL = 'Brnd_Name'    # Column for Brand Name
N_TOP_MANUFACTURERS_TO_SHOW = 10 # Number of top manufacturers to show individually

if MANUFACTURER_NAME_COL not in medicare_d_spending.columns or BRAND_NAME_COL not in medicare_d_spending.columns:
    print(f"Error: Required columns ('{MANUFACTURER_NAME_COL}' or '{BRAND_NAME_COL}') not found in your CSV.")
    print("Available columns are:", medicare_d_spending.columns.tolist())
    exit()

spending_cols_raw = [f'Tot_Spndng_{year}' for year in years]
missing_cols = [col for col in spending_cols_raw if col not in medicare_d_spending.columns]
if missing_cols:
    print(f"Error: Missing spending columns in your CSV: {missing_cols}")
    print("Please ensure your CSV has columns like 'Tot_Spndng_2018', etc. for your specified years.")
    exit()

# --- 3. Clean and convert spending columns to numeric ---
print("\nCleaning and converting spending columns to numeric...")
for col in spending_cols_raw:
    if col in medicare_d_spending.columns:
        medicare_d_spending[col] = medicare_d_spending[col].astype(str).str.replace(r'[$,]', '', regex=True)
        medicare_d_spending[col] = pd.to_numeric(medicare_d_spending[col], errors='coerce')
        medicare_d_spending[col] = medicare_d_spending[col].fillna(0)
    else:
        print(f"Warning: Spending column '{col}' not found, skipping conversion.")

print("Dtypes of spending columns after cleaning:")
print(medicare_d_spending[spending_cols_raw].dtypes)


# --- 4. Calculate Overall Total Spending for Each Manufacturer Across All Years ---
manufacturer_overall_spending = medicare_d_spending.groupby(MANUFACTURER_NAME_COL)[spending_cols_raw].sum().sum(axis=1)

# --- 5. Calculate Number of Brands per Manufacturer ---
# Group by manufacturer and count unique brand names
manufacturer_brand_counts = medicare_d_spending.groupby(MANUFACTURER_NAME_COL)[BRAND_NAME_COL].nunique()

# --- 6. Filter out "OVERALL" or similar unwanted names BEFORE identifying top N ---
# This line will remove any row where the manufacturer name contains 'OVERALL' (case-insensitive check)
manufacturer_overall_spending = manufacturer_overall_spending[
    ~manufacturer_overall_spending.index.str.contains('OVERALL', case=False, na=False)
]
manufacturer_brand_counts = manufacturer_brand_counts[
    ~manufacturer_brand_counts.index.str.contains('OVERALL', case=False, na=False)
]


# --- 7. Identify Top N Manufacturers (by spending) and get their brand counts ---
top_N_manufacturers_spending = manufacturer_overall_spending.nlargest(N_TOP_MANUFACTURERS_TO_SHOW)
top_N_manufacturer_names = top_N_manufacturers_spending.index.tolist()

# Get brand counts for the identified top N manufacturers
top_N_brands_count = manufacturer_brand_counts.loc[top_N_manufacturer_names].tolist()


# --- 8. Calculate total spending and total brands for all other manufacturers ---
other_manufacturers_spending_value = manufacturer_overall_spending[
    ~manufacturer_overall_spending.index.isin(top_N_manufacturer_names)
].sum()

other_manufacturers_brands_count = manufacturer_brand_counts[
    ~manufacturer_brand_counts.index.isin(top_N_manufacturer_names)
].sum()


# --- 9. Prepare data for the combined pie chart ---
pie_data_manufacturers = pd.DataFrame({
    'Manufacturer': top_N_manufacturer_names + ['Other Manufacturers'],
    'Overall_Spending': top_N_manufacturers_spending.tolist() + [other_manufacturers_spending_value],
    'Brand_Count': top_N_brands_count + [other_manufacturers_brands_count] # Add brand counts
})

# Convert spending to millions for better labels
pie_data_manufacturers['Overall_Spending_Millions'] = pie_data_manufacturers['Overall_Spending'] / 1_000_000

print(f"\nOverall Spending & Brand Count Data for Top {N_TOP_MANUFACTURERS_TO_SHOW} Manufacturers and Others:")
print(pie_data_manufacturers)

# --- 10. Plot the Pie Chart ---
plt.figure(figsize=(14, 14)) # Increased figure size for more space for labels

# Define colors for the slices
colors = sns.color_palette('tab10', n_colors=N_TOP_MANUFACTURERS_TO_SHOW) + ['grey']

# Explode the top manufacturer's slice slightly, or none for the others
explode_effect = [0.05] * N_TOP_MANUFACTURERS_TO_SHOW + [0]

wedges, texts, autotexts = plt.pie(
    pie_data_manufacturers['Overall_Spending_Millions'],
    explode=explode_effect,
    labels=None, # We will add custom labels to avoid clutter
    colors=colors,
    autopct='%1.1f%%', # Format percentages to one decimal place
    shadow=True,
    startangle=90,
    pctdistance=0.85 # Distance of percentage labels from center
)

plt.axis('equal') # Ensures the pie chart is circular.
plt.title(f'Overall Spending & Brand Count Distribution by Manufacturer (2018-2022)\nTop {N_TOP_MANUFACTURERS_TO_SHOW} Manufacturers vs. Others', fontsize=16)

# Add custom labels with names, percentages, and brand counts, positioned outside for clarity
total_spending = pie_data_manufacturers['Overall_Spending_Millions'].sum()
for i, (wedge, row_data) in enumerate(zip(wedges, pie_data_manufacturers.itertuples())):
    manufacturer_name = row_data.Manufacturer
    spending_value = row_data.Overall_Spending_Millions
    brand_count = row_data.Brand_Count

    angle = (wedge.theta2 - wedge.theta1) / 2. + wedge.theta1
    x = 1.25 * wedge.r * np.cos(np.deg2rad(angle)) # Further adjust multiplier for label distance
    y = 1.25 * wedge.r * np.sin(np.deg2rad(angle)) # Further adjust multiplier for label distance

    # Custom label format: Name\n(Percentage)\n(X Brands)
    label_text = f"{manufacturer_name}\n({spending_value/total_spending:.1%})\n({int(brand_count)} Brands)"
    plt.text(x, y, label_text, ha='center', va='center', fontsize=9, color='black')

plt.tight_layout()
plt.show()

print(f"\n**Note:** While this pie chart provides detailed information, displaying names, percentages, AND brand counts for {N_TOP_MANUFACTURERS_TO_SHOW} individual manufacturers (plus 'Other') can make it quite dense and hard to read, especially for smaller slices. For a clearer comparison of individual manufacturer spending and brand counts, a **horizontal bar chart** (perhaps with two bars per manufacturer for spending and brands, or a stacked bar chart) would typically be more effective.")
```
